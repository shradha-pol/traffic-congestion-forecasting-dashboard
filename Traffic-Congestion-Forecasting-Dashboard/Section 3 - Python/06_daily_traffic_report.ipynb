# Question 6: Automate daily traffic summary reports with Python scripts

# Step 1 - Import Pandas

```python
import pandas as pd
```

# Step 2 - Load Dataset

```python
df = pd.read_csv("traffic_data_full.csv")
```

# Step 3 - Convert Date Column

```python
df['Date'] = pd.to_datetime(df['Date'])
```

# Step 4 - Create Daily Traffic Summary

```python
daily_summary = df.groupby('Date').agg({
    'Traffic_Volume': 'sum',
    'Avg_Speed_kmph': 'mean',
    'Congestion_Ratio': 'mean'
}).reset_index()
```

# Step 5 - Rename Columns

```python
daily_summary.columns = [
    'Date',
    'Total_Traffic',
    'Average_Speed',
    'Average_Congestion'
]
```

# Step 6 - Display Report

```python
print(daily_summary.head())
```

# Step 7 - Export Report to CSV

```python
daily_summary.to_csv(
    "daily_traffic_summary.csv",
    index=False
)
```
