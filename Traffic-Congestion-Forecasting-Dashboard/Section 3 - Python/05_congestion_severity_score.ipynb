# Question 5: Write custom functions to calculate congestion severity scores

# Step 1 - Import Pandas

```python
import pandas as pd
```

# Step 2 - Load Dataset

```python
df = pd.read_csv("traffic_data_full.csv")
```

# Step 3 - Create Custom Function

```python
def congestion_severity(score):

    if score < 2:
        return "Low"

    elif score < 4:
        return "Medium"

    elif score < 6:
        return "High"

    else:
        return "Severe"
```

# Step 4 - Apply Function on Congestion Ratio

```python
df['Congestion_Severity'] = df['Congestion_Ratio'].apply(congestion_severity)
```

# Step 5 - Display Results

```python
print(df[['Congestion_Ratio', 'Congestion_Severity']].head())
```
