# Question 4: Use correlation analysis to check impact of weather on congestion

# Step 1 - Import Required Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Step 2 - Load Dataset

```python
df = pd.read_csv("traffic_data_full.csv")
```

# Step 3 - Select Numerical Columns

```python
correlation_data = df[[
    'Weather_Severity',
    'Traffic_Volume',
    'Avg_Speed_kmph',
    'Congestion_Ratio'
]]
```

# Step 4 - Calculate Correlation Matrix

```python
correlation_matrix = correlation_data.corr()

print(correlation_matrix)
```

# Step 5 - Visualize Correlation Heatmap

```python
plt.figure(figsize=(9,7))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    linewidths=0.5
)

plt.title(
    "Weather vs Traffic Congestion Correlation",
    pad=20
)

plt.xticks(rotation=15)
plt.yticks(rotation=0)

plt.tight_layout()

plt.show()
```
