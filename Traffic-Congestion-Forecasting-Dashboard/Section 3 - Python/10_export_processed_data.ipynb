# Question 10: Export processed datasets into CSV/Excel for Data Science stage

# Step 1 - Import Pandas

```python
import pandas as pd
```

# Step 2 - Load Processed Dataset

```python
df = pd.read_csv(
    "processed_traffic_data.csv"
)
```

```python
# Format Hour_Time column
df['Hour_Time'] = (
    df['Hour_Time']
    .astype(str)
    .str.split()
    .str[-1]
)
```

# Step 3 - Export Dataset to CSV

```python
df.to_csv(
    "traffic_data_for_datascience.csv",
    index=False
)
```

# Step 4 - Export Dataset to Excel

```python
df.to_excel(
    "traffic_data_for_datascience.xlsx",
    index=False
)
```
