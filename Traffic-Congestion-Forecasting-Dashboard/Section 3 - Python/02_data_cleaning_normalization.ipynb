# Question 2: Clean and normalize data using Python functions

# Step 1 - Import Libraries

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

# Step 2 - Load Dataset

```python
df = pd.read_csv("traffic_data_full.csv")
```

# Step 3 - Check Missing Values

```python
print(df.isnull().sum())
```

# Step 4 - Remove Duplicate Rows

```python
df = df.drop_duplicates()
```

# Step 5 - Convert Date and Timestamp Columns

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

# Step 6 - Standardize Text Columns

```python
df['Vehicle_Type'] = df['Vehicle_Type'].str.strip().str.title()
df['Weather'] = df['Weather'].str.strip().str.title()
```

# Step 7 - Normalize Numerical Columns

```python
scaler = MinMaxScaler()

df[['Traffic_Volume', 'Avg_Speed_kmph', 'Congestion_Ratio']] = scaler.fit_transform(
    df[['Traffic_Volume', 'Avg_Speed_kmph', 'Congestion_Ratio']]
)
```
