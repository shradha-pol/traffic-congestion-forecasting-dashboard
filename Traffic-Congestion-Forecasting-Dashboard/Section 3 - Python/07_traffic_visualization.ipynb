# Question 7: Visualize traffic flow using Matplotlib/Seaborn

# Step 1 - Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Step 2 - Load Dataset

```python
df = pd.read_csv("traffic_data_full.csv")
```

# Step 3 - Create Traffic Volume by Hour Plot

```python
plt.figure(figsize=(10,6))

sns.lineplot(
    data=df,
    x='Hour',
    y='Traffic_Volume',
    marker='o'
)

plt.title("Traffic Volume by Hour", fontsize=14)
plt.xlabel("Hour", fontsize=12)
plt.ylabel("Traffic Volume", fontsize=12)

plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

# Step 4 - Create Vehicle Type Distribution Chart

```python
plt.figure(figsize=(8,5))

ax = sns.countplot(
    data=df,
    x='Vehicle_Type',
    palette='Set2'
)

# Add count labels
for container in ax.containers:
    ax.bar_label(container)

plt.title("Vehicle Type Distribution", fontsize=14)
plt.xlabel("Vehicle Type", fontsize=12)
plt.ylabel("Count", fontsize=12)

plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

# Step 5 - Create Congestion Ratio Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['Congestion_Ratio'],
    bins=20,
    kde=True,
    color='skyblue',
    alpha=0.7
)

plt.title("Congestion Ratio Distribution", fontsize=14)
plt.xlabel("Congestion Ratio", fontsize=12)
plt.ylabel("Count", fontsize=12)

plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
