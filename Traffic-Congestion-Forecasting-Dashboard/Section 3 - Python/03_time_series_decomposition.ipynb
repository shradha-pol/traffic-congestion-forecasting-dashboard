# Question 3: Perform time-series decomposition (trend, seasonality, residuals)

# Step 1 - Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

# Step 2 - Load Dataset

```python
df = pd.read_csv("traffic_data_full.csv")
```

# Step 3 - Convert Timestamp Column

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

# Step 4 - Set Timestamp as Index

```python
df.set_index('Timestamp', inplace=True)
```

# Step 5 - Select Traffic Volume Data

```python
traffic_series = df.groupby('Timestamp')['Traffic_Volume'].sum()
```

# Step 6 - Perform Time-Series Decomposition

```python
decomposition = seasonal_decompose(
    traffic_series,
    model='additive',
    period=24
)
```

# Step 7 - Plot Decomposition Results

```python
# Plot decomposition results
fig = decomposition.plot()

# Increase graph size
fig.set_size_inches(14, 10)

# Rotate x-axis labels
for ax in fig.axes:
    ax.tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout()

# Display graph
plt.show()
```
