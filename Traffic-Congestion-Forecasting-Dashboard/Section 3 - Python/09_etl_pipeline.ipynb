# Question 9: Build modular Python scripts for ETL (Extract, Transform, Load)

# Step 1 - Import Required Library

```python
import pandas as pd
```

# Step 2 - Create Extract Function

```python
def extract_data(file_path):

    df = pd.read_csv(file_path)

    print("Data Extracted Successfully")

    return df
```

# Step 3 - Execute Extract Function

```python
traffic_df = extract_data(
    "traffic_data_full.csv"
)
```

# Step 4 - Create Transform Function

```python
def transform_data(df):

    # Remove duplicate rows
    df = df.drop_duplicates()

    # Convert timestamp column
    df['Timestamp'] = pd.to_datetime(
        df['Timestamp']
    )

    # Standardize vehicle type names
    df['Vehicle_Type'] = (
        df['Vehicle_Type']
        .str.strip()
        .str.title()
    )

    print("Data Transformed Successfully")

    return df
```

# Step 5 - Execute Transform Function

```python
traffic_df = transform_data(traffic_df)
```

# Step 6 - Create Load Function

```python
def load_data(df, output_file):

    df.to_csv(output_file, index=False)

    print("Data Loaded Successfully")
```

# Step 7 - Execute Load Function

```python
load_data(
    traffic_df,
    "processed_traffic_data.csv"
)
```
