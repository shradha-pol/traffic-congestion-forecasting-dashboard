# Question 8: Conduct Sensitivity Analysis on Traffic Volume vs Weather Conditions

# Step 1 - Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Step 2 - Load Dataset

```python
df = pd.read_csv("traffic_data_for_datascience.csv")
```

# Step 3 - Check Weather Column

```python
print(df['Weather'].unique())
```

# Step 4 - Calculate Average Traffic by Weather

```python
weather_analysis = df.groupby('Weather')['Traffic_Volume'].mean().reset_index()
```

# Step 5 - View Weather Analysis Data

```python
print(weather_analysis)
```

# Step 6 - Plot Sensitivity Analysis Chart

```python
plt.figure(figsize=(10,6))

colors = ['skyblue', 'gray', 'orange', 'green']

bars = plt.bar(
    weather_analysis['Weather'],
    weather_analysis['Traffic_Volume'],
    color=colors
)

plt.xlabel("Weather Condition")
plt.ylabel("Average Traffic Volume")

plt.title("Traffic Volume vs Weather Conditions")

for bar in bars:
    height = bar.get_height()

    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 0.5,
        f'{height:.1f}',
        ha='center'
    )


plt.show()
```
