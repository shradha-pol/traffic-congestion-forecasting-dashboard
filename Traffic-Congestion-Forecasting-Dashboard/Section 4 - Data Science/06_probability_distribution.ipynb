# Question 6: Calculate Congestion Probability Distributions

# Step 1 - Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Step 2 - Load Dataset

```python
df = pd.read_csv("traffic_data_for_datascience.csv")
```

# Step 3 - Create Traffic Clusters 

```python
def classify_traffic(volume):

    if volume < 100:
        return 'Low Traffic'

    elif volume < 200:
        return 'Medium Traffic'

    else:
        return 'High Traffic'
```

# Step 4 - Apply Traffic Classification

```python
df['Traffic_Cluster'] = df['Traffic_Volume'].apply(classify_traffic)
```

# Step 5 - Calculate Probability Distribution

```python
probability_distribution = df['Traffic_Cluster'].value_counts(normalize=True)
```

# Step 6 - Show Probability Values

```python
print(probability_distribution)
```

# Step 7 - Convert to Percentage

```python
probability_percentage = probability_distribution * 100

print(probability_percentage)
```

# Step 8 - Visualize Probability Distribution

```python
plt.figure(figsize=(8,5))

probability_percentage.plot(
    kind='bar'
)

plt.xlabel("Traffic Cluster")
plt.ylabel("Probability Percentage")

plt.title("Traffic Congestion Probability Distribution")

for i, value in enumerate(probability_percentage):
    plt.text(i, value + 1, f'{value:.1f}%', ha='center')

plt.show()
```
