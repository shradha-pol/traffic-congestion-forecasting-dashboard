# Question 5: Perform Clustering Using Rule-Based Thresholds (Without Machine Learning)

# Step 1 - Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Step 2 - Load Dataset

```python
df = pd.read_csv("traffic_data_for_datascience.csv")
```

# Step 3 - View Traffic Volume Statistics

```python
print(df['Traffic_Volume'].describe())
```

# Step 4 - Create Rule-Based Clusters

```python
def classify_traffic(volume):

    if volume < 100:
        return 'Low Traffic'

    elif volume < 200:
        return 'Medium Traffic'

    else:
        return 'High Traffic'
```

# Step 5 - Apply Clustering Rules

```python
df['Traffic_Cluster'] = df['Traffic_Volume'].apply(classify_traffic)
```

# Step 6 - View Clustered Data

```python
df[['Traffic_Volume',
    'Traffic_Cluster']].head(10)
```

# Step 7 - Count Cluster Categories

```python
cluster_counts = df['Traffic_Cluster'].value_counts()

print(cluster_counts)
```

# Step 8 - Visualize Clusters

```python
plt.figure(figsize=(8,5))

cluster_counts.plot(
    kind='bar'
)

plt.xlabel("Traffic Cluster")
plt.ylabel("Count")

plt.title("Traffic Congestion Clusters")

plt.show()
```
