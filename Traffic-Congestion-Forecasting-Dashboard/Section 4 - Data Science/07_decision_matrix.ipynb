# Question 7: Build Decision Matrices for Traffic Management Interventions

# Step 1 - Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Step 2 - Load Dataset

```python
df = pd.read_csv("traffic_data_for_datascience.csv")
```

# Step 3 - Create Traffic Severity Categories

```python
def classify_traffic(volume):

    if volume < 100:
        return 'Low Traffic'

    elif volume < 200:
        return 'Medium Traffic'

    else:
        return 'High Traffic'
```

# Step 4 - Apply Traffic Classification

```python
df['Traffic_Cluster'] = df['Traffic_Volume'].apply(classify_traffic)
```

# Step 5 - Create Decision Matrix Rules

```python
def traffic_intervention(cluster):

    if cluster == 'Low Traffic':
        return 'Normal Monitoring'

    elif cluster == 'Medium Traffic':
        return 'Signal Optimization'

    else:
        return 'Emergency Traffic Control'
```

# Step 6 - Apply Decision Matrix

```python
df['Traffic_Action'] = df['Traffic_Cluster'].apply(traffic_intervention)
```

# Step 7 - View Decision Matrix Data

```python
df[['Traffic_Volume',
    'Traffic_Cluster',
    'Traffic_Action']].head(10)
```

# Step 8 - Count Recommended Actions

```python
action_counts = df['Traffic_Action'].value_counts()

print(action_counts)
```

# Step 9 - Visualize Traffic Management Actions

```python
plt.figure(figsize=(8,5))

action_counts.plot(kind='bar')

plt.xlabel("Traffic Management Action")
plt.ylabel("Count")

plt.title("Traffic Management Decision Matrix")

plt.show()
```
