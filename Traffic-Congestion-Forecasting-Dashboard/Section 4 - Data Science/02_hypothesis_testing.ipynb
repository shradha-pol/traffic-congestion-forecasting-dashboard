# Question 2: Conduct Hypothesis Testing “Is congestion significantly higher on weekdays?”


# Step 1 - Import Libraries

```python
import pandas as pd
from scipy.stats import ttest_ind
```

# Step 2 - Load Dataset

```python
df = pd.read_csv("traffic_data_for_datascience.csv")
```

# Step 3 - Check Columns

```python
df.columns
```

# Step 4 - Convert Date Column to Datetime

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

# Step 5 - Create Day Type Column

```python
df['Day_Type'] = df['Timestamp'].dt.dayofweek
```

# Step 6 - Separate Weekday and Weekend Data

```python
weekday_data = df[df['Day_Type'] < 5]['Traffic_Volume']

weekend_data = df[df['Day_Type'] >= 5]['Traffic_Volume']
```

# Step 7 - Perform T-Test

```python
t_stat, p_value = ttest_ind(weekday_data, weekend_data)
```

# Step 8 - Show Results

```python
print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

# Step 9 - Interpret Result

```python
if p_value < 0.05:
    print("Congestion is significantly higher on weekdays.")
else:
    print("No significant difference in congestion.")
```
