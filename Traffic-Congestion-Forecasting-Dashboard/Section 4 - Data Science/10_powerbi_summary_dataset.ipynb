# Question 10: Prepare Summarized Datasets for Power BI Visualization

# Step 1 - Import Libraries

```python
import pandas as pd
```

# Step 2 - Load Dataset

```python
df = pd.read_csv("traffic_data_for_datascience.csv")
```

# Step 3 - Create Traffic Clusters

```python
def classify_traffic(volume):

    if volume < 100:
        return 'Low Traffic'

    elif volume < 200:
        return 'Medium Traffic'

    else:
        return 'High Traffic'
```

# Step 4 - Apply Traffic Clustering

```python
df['Traffic_Cluster'] = df['Traffic_Volume'].apply(classify_traffic)
```

# Step 5 - Create Weather Severity Scores

```python
Weather = {

    'Clear': 1,

    'Cloudy': 2,

    'Rain': 3,

    'Fog': 4

}

df['Weather_Severity'] = df['Weather'].map(Weather)
```

# Step 6 - Normalize Traffic Volume

```python
df['Normalized_Traffic'] = (

    (df['Traffic_Volume'] - df['Traffic_Volume'].min())

    /

    (df['Traffic_Volume'].max() - df['Traffic_Volume'].min())

)
```

# Step 7 - Normalize Average Speed

```python
df['Normalized_Speed'] = (

    1 -

    (

        (df['Avg_Speed_kmph'] - df['Avg_Speed_kmph'].min())

        /

        (df['Avg_Speed_kmph'].max() - df['Avg_Speed_kmph'].min())

    )

)
```

# Step 8 - Normalize Weather Severity

```python
df['Normalized_Weather'] = (

    (df['Weather_Severity'] - df['Weather_Severity'].min())

    /

    (df['Weather_Severity'].max() - df['Weather_Severity'].min())

)
```

# Step 9 - Create Congestion Index

```python
df['Congestion_Index'] = (

    df['Normalized_Traffic'] +

    df['Normalized_Speed'] +

    df['Normalized_Weather']

) / 3
```

# Step 11 - Export Final Power BI Dataset

```python
df.to_csv(

    "powerbi_final_dataset.csv",

    index=False
)
```
