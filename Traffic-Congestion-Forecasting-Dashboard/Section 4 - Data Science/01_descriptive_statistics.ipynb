# Question 1: Perform Descriptive Statistics on Traffic Volume

# Step 1 - Import Libraries

```python
import pandas as pd
import numpy as np
```

# Step 2 - Load Dataset

```python
df = pd.read_csv("traffic_data_for_datascience.csv")
```

# Step 3 - Check Dataset

```python
df.head()
```

# Step 4 - Check Column Names

```python
df.columns
```

# Step 5 - Perform Descriptive Statistics

```python
traffic_stats = df['Traffic_Volume'].describe()

print(traffic_stats)
```

# Step 6 - Calculate Variance Separately

```python
variance = df['Traffic_Volume'].var()

print("Variance:", variance)
```

# Step 7 - Calculate Median Separately

```python
median = df['Traffic_Volume'].median()

print("Median:", median)
```

# Step 8 - Full Statistics Table

```python
stats_summary = {
    "Mean": df['Traffic_Volume'].mean(),
    "Median": df['Traffic_Volume'].median(),
    "Variance": df['Traffic_Volume'].var(),
    "Standard Deviation": df['Traffic_Volume'].std(),
    "Minimum": df['Traffic_Volume'].min(),
    "Maximum": df['Traffic_Volume'].max()
}

for key, value in stats_summary.items():
    print(f"{key}: {value}")
```
