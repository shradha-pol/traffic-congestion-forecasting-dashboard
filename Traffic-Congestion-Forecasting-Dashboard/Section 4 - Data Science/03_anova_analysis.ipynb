# Question 3: Use ANOVA to Compare Congestion Across Multiple Road Segments

# Step 1 - Import Libraries

```python
import pandas as pd
from scipy.stats import f_oneway
```

# Step 2 - Load Dataset

```python
df = pd.read_csv("traffic_data_for_datascience.csv")
```

# Step 3 - Check Column Names

```python
df.columns
```

# Step 4 - Check Unique Road Segments

```python
df['Road_Segment'].unique()
```

# Step 5 - Separate Traffic Data by Road Segment

```python
segment1 = df[df['Road_Segment'] == 'Eastern Express Highway']['Traffic_Volume']

segment2 = df[df['Road_Segment'] == 'Bandra Link Road']['Traffic_Volume']

segment3 = df[df['Road_Segment'] == 'Mumbai-Pune Highway']['Traffic_Volume']
```

# Step 6 - Perform ANOVA Test

```python
f_stat, p_value = f_oneway(segment1, segment2, segment3)
```

# Step 7 - Show Results

```python
print("F-Statistic:", f_stat)
print("P-Value:", p_value)
```

# Step 8 - Interpret Result

```python
if p_value < 0.05:
    print("Traffic congestion differs significantly across road segments.")
else:
    print("No significant difference across road segments.")
```
